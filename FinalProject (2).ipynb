```python
import pandas as pd
import numpy as np

df = pd.read_csv("NeighborhoodsDataSet.csv")

numeric_cols = [
    "Population (2010)",
    "Street Density (st. mi/area sq. mi)",
    "Total Street Miles",
    "# Sets of Steps",
    "Miles of Major Roads"
]

for col in numeric_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "")
        .replace("None", np.nan)
        .astype(float)
    )

df["Norm_Population"] = (df["Population (2010)"] - df["Population (2010)"].min()) / (df["Population (2010)"].max() - df["Population (2010)"].min())
df["Norm_Street_Density"] = (df["Street Density (st. mi/area sq. mi)"] - df["Street Density (st. mi/area sq. mi)"].min()) / (df["Street Density (st. mi/area sq. mi)"].max() - df["Street Density (st. mi/area sq. mi)"].min())
df["Norm_Total_Street_Miles"] = (df["Total Street Miles"] - df["Total Street Miles"].min()) / (df["Total Street Miles"].max() - df["Total Street Miles"].min())
df["Norm_Num_Steps"] = (df["# Sets of Steps"] - df["# Sets of Steps"].min()) / (df["# Sets of Steps"].max() - df["# Sets of Steps"].min())
df["Norm_Major_Road_Miles"] = (df["Miles of Major Roads"] - df["Miles of Major Roads"].min()) / (df["Miles of Major Roads"].max() - df["Miles of Major Roads"].min())

df["Bestness_Score"] = (
    0.30 * df["Norm_Population"] +
    0.25 * df["Norm_Street_Density"] +
    0.20 * df["Norm_Total_Street_Miles"] +
    0.15 * df["Norm_Num_Steps"] +
    0.10 * df["Norm_Major_Road_Miles"]
)

ranked = df[["Neighborhood", "Bestness_Score"]].sort_values("Bestness_Score", ascending=False)

ranked
```

```python
plt.figure(figsize=(10,6))
sns.histplot(ranked["Bestness_Score"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Bestness Scores Across Pittsburgh Neighborhoods")
plt.xlabel("Bestness Score")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.barplot(
    data=ranked.head(15),
    x="Neighborhood",
    y="Bestness_Score",
    palette="viridis"
)
plt.xticks(rotation=75, ha="right")
plt.title("Top 15 Pittsburgh Neighborhoods ‚Äî Bestness Score")
plt.ylabel("Bestness Score")
plt.xlabel("Neighborhood")
plt.tight_layout()
plt.show()
```

```python
# Identify the top neighborhood
best_neighborhood = ranked.iloc[0]["Neighborhood"]
best_score = ranked.iloc[0]["Bestness_Score"]

print("üèÜ BEST NEIGHBORHOOD SUMMARY\n")
print(f"The best neighborhood in Pittsburgh according to our Bestness Score model is **{best_neighborhood}**,")
print(f"with a Bestness Score of **{best_score:.4f}**."
```
